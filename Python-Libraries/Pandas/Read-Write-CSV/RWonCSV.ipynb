```python
def fixed_currency(cur):
    if cur == "$$" or cur == 'Dollars':
        return 'USD'
    return cur


df_financials = pd.read_excel("Data-File/movies_db.xlsx", "financials", converters={
    'currency' : fixed_currency
})
df_financials.head(5)
```

```python
df_merge = pd.merge(df_movi, df_financials, on="movie_id")
df_merge.head(5)
```

```python
df_merge.to_excel("Movie_merged.xlsx", sheet_name="Merged", index=False)
```
